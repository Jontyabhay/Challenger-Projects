### 1. Import Packages

```python
import pandas as pd
import numpy as np
```

```python
import pandas_datareader as web
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

### 2. Load the adjusted closings for the top 5 tech stocks.

```python
symbols = ['MSFT','AMZN','AAPL','GOOG','FB']
start_date = '2019-01-01'
end_date = '2019-07-01'
stock_data = web.get_data_yahoo(symbols, start_date, end_date)
print(stock_data,stock_data['Adj Close'])
```

### 3. Plot the adjusted closing prices over time.

```python
plt.plot(stock_data['Adj Close'])
plt.xlabel("Date")
plt.ylabel("Adjusted Closing Price Over Time")
plt.title("Tech Stocks Adjusted Price")
plt.legend(symbols)
plt.show()
```

```python
percent_return = stock_data['Adj Close'].pct_change()
plt.plot(percent_return)
plt.xlabel("Date")
plt.ylabel("Adjusted Percent Price Over Time")
plt.title("Tech Stocks Over Time")
plt.legend(symbols)
plt.figure(figsize=(16,9))
plt.show()
```

```python
fig = plt.figure(figsize=(15,15))

ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)

ax1.plot(stock_data['Adj Close']['MSFT'].pct_change())
ax1.set_title("Microsoft")
ax2.plot(stock_data['Adj Close']['AMZN'].pct_change())
ax2.set_title('Amazon')
ax3.plot(stock_data['Adj Close']['AAPL'].pct_change())
ax3.set_title("Apple")
ax4.plot(stock_data['Adj Close']['GOOG'].pct_change())
ax4.set_title('Google')
ax5.plot(stock_data['Adj Close']['FB'].pct_change())
ax5.set_title('Facebook')

plt.tight_layout()
plt.show()
```

### 6. Calculate and plot the mean of each tech stock's daily simple rate of return

```python
mean_rate = stock_data['Adj Close'].pct_change().mean()
mean_rate

height = []
for key in mean_rate.keys():
    height.append(mean_rate[key])
height

x_pos = np.arange(len(mean_rate.keys()))
x_pos

plt.bar(x_pos, height)
plt.xticks(x_pos, mean_rate.keys())
plt.xlabel("Tech stocks")
plt.ylabel("mean")
plt.title("Daily Mean")
plt.show()
```

### 7. Calculate and plot the variance.

```python
var_rate = stock_data['Adj Close'].pct_change().var()
var_rate

height = []
for key in var_rate.keys():
    height.append(var_rate[key])
height

x_pos = np.arange(len(var_rate.keys()))
x_pos

plt.bar(x_pos, height)
plt.xticks(x_pos, var_rate.keys())
plt.xlabel("Tech Stocks")
plt.ylabel("variance")
plt.title("Daily Variance")
plt.show()
```

### 8. Calculate and plot the standard deviation

```python
std_rate = stock_data['Adj Close'].pct_change().std()
std_rate

height = []
for key in std_rate.keys():
    height.append(std_rate[key])
height

x_pos = np.arange(len(std_rate.keys()))
x_pos

plt.bar(x_pos,height)
plt.xticks(x_pos, std_rate.keys())
plt.xlabel("Tech Stocks")
plt.ylabel("std")
plt.title("Daily Standard Deviation")
plt.show()
```
